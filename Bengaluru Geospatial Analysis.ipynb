```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
data = pd.read_csv(r'C:\Users\shrey\Downloads\Banglore_traffic_Dataset.csv')
data.columns
```

```python
data.head(10)
```

```python
# Area locations (latitude and longitude for each area)
area_locations = {
    'Indiranagar': [12.9716, 77.6412],
    'Whitefield': [12.9719, 77.7500],
    'Koramangala': [12.9352, 77.6245],
    'MG Road': [12.9716, 77.6075],
    'BTM Layout': [12.9257, 77.6070],
    # We can add more area names and coordinates if needed
}

# Create a base map of Bangalore
bangalore_map = folium.Map(location=[12.9716, 77.5946], zoom_start=12)

# Heatmap data assignment
heat_data = []

# Loop through the dataset and prepare the heat data
for index, row in data.iterrows():
    area_name = row['Area Name']  # Make sure this matches your dataset column name
    if area_name in area_locations:
        lat, lon = area_locations[area_name]
        congestion_level = row['Congestion Level'] / 100  # Normalize congestion level (assuming it's a percentage)
        heat_data.append([lat, lon, congestion_level])

# Map populated check
if heat_data:
    HeatMap(heat_data).add_to(bangalore_map)
    display(bangalore_map)
else:
    print("No heatmap data available! Please check your dataset and area mappings.")
```
